```python
with open('testinput.txt') as f:
    data = f.read()
print (data)
```

```python
ponctuation="’-°.,;'':/_—?!…"
for p in ponctuation:
    data = data.replace(p," ")
print (data)
```

```python
nbLigne = data.count("\n")
print (nbLigne)
```

```python
data = data.lower()
print (data)
```

```python
mots = data.split()
nbMots = len(mots)
print (nbMots)
```

```python
motUnique= set(mots)
motAsupp=[ m for m in motUnique if len(m)<7]
for mm in motAsupp:
    motUnique.remove(mm)

nbMotUnique=len(motUnique)
print(nbMotUnique)
```

```python
dicMotUnique={}
for z in motUnique:
    dicMotUnique[z]=data.count(z)
print(len(dicMotUnique))
```

```python
def motPlusUtilise(dico):
    maxMots = []
    for k, v in dico.items():
        maxMots.append((v, k))
    maxMots.sort(reverse=True)
    return maxMots
```

```python
top10 = motPlusUtilise(dicMotUnique)
for k, v in top10[:10]:
    print(v, k)
```
